## Import Data from the files and consolidate to a single file

### Import python libraries

```python
import numpy as np
import pandas as pd
import gc
```

```python
pd.set_option('display.max_columns', None)  # Show all columns
pd.set_option('display.max_rows', None)     # Show all rows
# Force pandas to display 2 decimal places (no scientific notation)
pd.options.display.float_format = '{:.2f}'.format
```

### Consolidate the data files 
This process will first generate the list of all files prior to 2019 and another list with files greater than 2019. The data will be consolidated into one file using pd.concat()

```python
pre_2019 = []
after_2019 = []
for year in range(2005, 2025):
    if year < 2019:
        pre_2019.append(f'neiss{year}.xlsx')
    else:   
        after_2019.append(f'neiss{year}.xlsx')
print(pre_2019)
print(after_2019)
```

Read each file in to a dataframe and do the below
1. Add the year column in the dataframe
2. Concatenate the data frame to a single file and save as a csv 
3. Create a pre 2019 dataframe and save it as a CSV
4. Create a dataset for the data after 2018 and save it as a CSV
5. Combine the pre 2019 and post 2018 datasets into a new CSV


```python
data_pre2019 = pd.DataFrame()
for file_name in pre_2019:
    file_path = f'../data/raw/{file_name}'
    df = pd.read_excel(file_path, engine='calamine')
    print(f'{file_name}: {df.shape}')
    #df['data_year'] = int(file_name[5:9])  # Extract year from filename
    df.insert(0, 'data_year',int(file_name[5:9])) 
    data_pre2019=pd.concat([data_pre2019, df], ignore_index=True)
print(f'Consolidated pre-2019 data shape: {data_pre2019.shape}')
#data_pre2019.to_csv('../data/raw/neiss_pre_2019.csv', index=False)  
data_pre2019.to_parquet('../data/raw/neiss_pre_2019..parquet')
```

```python
data_after2019 = pd.DataFrame()
for file_name in after_2019:
    file_path = f'../data/raw/{file_name}'
    df = pd.read_excel(file_path, engine='calamine')
    print(f'{file_name}: {df.shape}')
    #df['data_year'] = int(file_name[5:9])  # Extract year from filename
    df.insert(0, 'data_year',int(file_name[5:9])) 
    data_after2019=pd.concat([data_after2019, df], ignore_index=True)
print(f'Consolidated after-2019 data shape: {data_after2019.shape}')
#data_after2019.to_csv('../data/raw/neiss_after_2019.csv', index=False)  
data_after2019.to_parquet('../data/raw/neiss_after_2019..parquet')
```

```python
combined_df = pd.concat([data_pre2019, data_after2019], sort=False, ignore_index=True)
print(f"Old shape: {data_pre2019.shape}")
print(f"New shape: {data_after2019.shape}")
print(f"Combined shape: {combined_df.shape}")
#combined_df.to_csv('../data/raw/neiss_original_raw_untouched_data.csv', index=False)
combined_df.to_parquet('../data/raw/neiss_original_raw_untouched_data.parquet')
```
